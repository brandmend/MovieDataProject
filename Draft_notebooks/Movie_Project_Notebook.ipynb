#### What movies have the highest ratings?

```python
ratings_total_weight = pd.read_sql_query('''
SELECT 
mr.movie_id
,mb.primary_title
,mb.original_title
,mb.genres
,mb.start_year
,mb.runtime_minutes
,averagerating
,numvotes
,(numvotes*averagerating) as total_weight

FROM movie_ratings mr
LEFT JOIN movie_basics mb
ON mr.movie_id = mb.movie_id

GROUP BY 1,2,3,4,5,6,7

ORDER BY total_weight DESC 


;

''',conn)

ratings_total_weight.info()
```

```python
tmdb_data['total_weight'] = tmdb_data['vote_average']*tmdb_data['vote_count']

tmdb_data.sort_values(by='total_weight',ascending=False,inplace=True)

tmdb_data.head(20)
```
